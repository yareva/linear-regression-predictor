```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# Simple data (X: hours studied, y: test score)
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 5, 4, 5])

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# !pip install matplotlib sklearn (sklearn is the old library)
# !pip install scikit-learn

# Predict
X_test = np.array([[6]])
#This creates a 2D Numpy array with the value 6.
# n_samples = 1 (just one student)
# n_features = 1 (just one featureL hours studied)
# So [[6]] is correct. [[x]] is common for a single data point in scikit-learn.

y_pred = model.predict(X_test)
# This uses the trained model to predict the output (test score) for the input X_test.
# model.predict() applies the learned regression equation: score = slope * 6 + intercept
# The result, y_pred, is an array containing the predicted score: array([5.2])

print(f"Predicted score for 6 hours of study: {y_pred[0]:.2f}")

# Plotting
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, model.predict(X), color='red', label='Best fit line')
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.title('Linear regression Example')
plt.legend()
plt.show()
```
